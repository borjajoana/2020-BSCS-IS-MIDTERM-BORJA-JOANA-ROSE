```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

worse = [855.2, 724.8, 356.9, 141.3, 124.3, 110.3, 79.7, 44.3, 26.9, 20.4]
moderate = [421.8, 497.7, 221.0, 98.9, 61.6, 50.5, 56.8, 30.5, 8.7, 10.0]
best = [82.1, 93.2, 41.5, 18.5, 11.7, 9.4, 10.7, 5.7, 5.2, 4.1]
plt.rc('figure', figsize=(25,10), facecolor="lavender")



x = np.arange(len(worse))  # the label locations
width = 0.3  # the width of the bars


fig, ax = plt.subplots()
rects1 = ax.bar(x + width/1, worse, width, label='Worse Case', color='green')
rects2 = ax.bar(x + width/3, moderate, width, label='Moderate Case', color='blue')
rects3 = ax.bar(x - width/2, best, width, label='Best Case', color='black')

def label(rects):
    
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')


label(rects1)
label(rects2)
label(rects3)


plt.title("IMPACT OF COVID-19 ON KEY PHILIPPINE ECONOMIC SECTORS", fontsize="35", color="black")
plt.ylabel('(in P billions)', fontsize='18')

plt.xticks([i + 0.30 for i in range(10)], ['Manufacuring', 'Wholesale and Retail Trade', 'Other Services', 'Financial Intermediation', 'Transportation, Storage and Communication', 'Agriculture, Forestry and Fishing', 'Real Estate, Renting and Business Activities', 'Electricity, Gas and Water ',  ' Mining and Quarrying', 'Construction'], rotation="90", fontsize="11")

plt.xticks(x, size="11")
plt.grid(linestyle='--', linewidth=0.3)
ax.legend(loc="best", shadow=True, fontsize="20", facecolor="lavender")

plt.show()

```
